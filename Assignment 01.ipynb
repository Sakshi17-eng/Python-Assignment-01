###2. Describe the role of predefined keywords in Python and provide examples of how they are used in a program
Answer:
Predefined keywords are the words that have special meaning and structure that has purpose within languages syntax.They are building blocks that tells how a python code is executed.
Some of the keywords are:
- Conditional Statements(if,else)


```python
temperature = 25

if temperature > 30:
    print("It's hot!")
else:
    print("It's cool.")
```

- Boolean values(true,false)

```python
is_sunny = True

if is_sunny:
    print("It's a bright day!")
else:
    print("It's not sunny.")
```

- Logical operators(and,or, not)



```python
has_coffee = True
has_pastry = False

if has_coffee and has_pastry:
    print("Enjoy your coffee and pastry!")
else:
    print("Sorry, nothing here for you right now.")
if has_coffee or has_pastry:
    print("Enjoy your coffee and pastry!")
else:
    print("Choose what is available right now")
```

###4. Discuss the different types of operators in Python and provide examples of how they are used.
Answer:
Operators are special symbols in pythons they performs operations on operands that are values , some of the operators are as follows:

- Arithmetic Operators:


1.   Addition(+) : adds two or more values
2.   Subtraction(-): subtracts two or more values
3. Multiplication(*): multiples two or more values
4. Division(/): divides two values
5. Modulas(%): returns the remainder after division



```python
a = 16
b = 4

print(f"a + b = {a + b}")
print(f"a - b = {a - b}")
print(f"a * b = {a * b}")
print(f"a / b = {a / b}")
print(f"a % b = {a % b}")
```

- Logical Operators:

 These operators perform logical operations on boolean values.


 1. and (AND):	Returns True if both operands are True = True and True =	True
 2. or	(OR):	Returns True if at least one operand is True = True or False=	True
 3. not	(NOT):	Returns True if the operand is False, and vice versa= not false=true



```python
p = True
q = False

print(f"p and q: {p and q}")
print(f"p or q: {p or q}")
print(f"not p: {not p}")
print(f"not q: {not q}")
```

 - Comparison Operators:

These operators compare the values of two operands and return a boolean result (True or False).

1. == :	Equal to -	Returns True if both operands are equal :	5 == 5	(True)
2. !=:	Not equal to -	Returns True if operands are not equal:	5 != 3	(True)
3. >:	Greater than -	Returns True if the left operand is greater than the right:	8 > 3	(True)
4. <:	Less than -	Returns True if the left operand is less than the right	:2 < 7	(True)
5. >=:	Greater than or equal to -	Returns True if the left operand is greater than or equal to the right:	5 >= 5	(True)
6. <=:	Less than or equal to -	Returns True if the left operand is less than or equal to the right:	4 <= 6	(True)


```python
x = 9
y = 12

print(f"x == y: {x == y}")
print(f"x != y: {x != y}")
print(f"x > y: {x > y}")
print(f"x < y: {x < y}")
print(f"x >= y: {x >= y}")
print(f"x <= y: {x <= y}")
```

###6. How do conditional statements work in Python? Illustrate with examples
Answer: In python,Conditional statements allows you to control the flow of your execution whether the certain conditions are true or false
- The primary conditional statements in Python are:

if statement: Executes a block of code if a specified condition is true.



else statement: Executes a block of code "if" all and conditions were false. You can have at most one else statement in a conditional structure.

```python
number = 11

if number % 2 == 0:
    print(f"{number} is an even number.")
else:
    print(f"{number} is an odd number.")
```
